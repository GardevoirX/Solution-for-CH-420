```python
"""

import necessary libraries

"""
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''

As an example: I set U/N as a constant value 10, and use N = 4, 10, 50 (U = 40, 100, 500) to compare the results.

necessary commands:
matplotlib

397  ./run
398  LS
399  ls
400  mv results.dat results_4.dat
401  ./run
402  mv results.dat results_10.dat
403  ./runs
404  mv results.dat results_50.dat

'''

file_name = ["../Source/result_4.dat", "../Source/result_10.dat", "../Source/result_50.dat"]
num_ls = [4, 10, 50]

fz = 14
for n, file in zip(num_ls, file_name):
    data = np.loadtxt(file)
    plt.plot(data[:,0], data[:,1], label = "N = %d" %n)
plt.xlabel("Energy level (U/N = 10)", fontsize = fz)
plt.ylabel("Probability", fontsize = fz)
plt.legend(fontsize = fz)
```

```python

file_name = ["../Source/result_4.dat", "../Source/result_10.dat", "../Source/result_50.dat"]
num_ls = [4, 10, 50]

fz = 14
for n, file in zip(num_ls, file_name):
    data = np.loadtxt(file)
    plt.plot(n*10-data[:,0], data[:,1], label = "Other oscallators in N = %d" %n)
plt.xlabel("Energy level U/N = 10", fontsize = fz)
plt.ylabel("Probability", fontsize = fz)
plt.legend(fontsize = fz-4)
```

```python
"""

Set N = 50

First run NVT and random choose beta = 1; get the Average Energy First Oscillator :0.57


Run NVE and let U ~= 0.57*N; get the Average Energy First Oscillator :0.58, U = 29


"""

file_name = ["../Source/results_nvt.dat", "../Source/results_nve.dat"]
num_ls = ["NVT", "NVE"]

fz = 14 
for n, file in zip(num_ls, file_name):
    data = np.loadtxt(file)
    if n == "NVT":
        plt.plot(data[:,0], data[:,1], label = n, color = "red")
    if n == "NVE":
        plt.scatter(data[:,0], data[:,1], label = n, color = "blue")
plt.xlabel("Energy level", fontsize = fz)
plt.ylabel("Probability", fontsize = fz)
plt.legend(fontsize = fz)
```
